```python
result = []
player_id = data['entry'].tolist()
player_name = data['player_name'].tolist()
all_df = []

for team in player_id:
    url_tid = 'https://fantasy.premierleague.com/api/entry/' + str(team) + '/history/' #inidividual address for each player
    p_id = requests.get(url_tid).text
    json_data_p_id = json.loads(p_id)
#     print(json.dumps(json_data_p, indent =2))  #prints JSON for every player
    for x in range(len(json_data_p_id['current'])):
        row = [team, 
               json_data_p_id['current'][x]['event'], 
               json_data_p_id['current'][x]['points'],
               json_data_p_id['current'][x]['total_points'],
               json_data_p_id['current'][x]['value'], 
               json_data_p_id['current'][x]['event_transfers'],
               json_data_p_id['current'][x]['points_on_bench'],
               json_data_p_id['current'][x]['event_transfers_cost']]
        new_DF = pd.DataFrame([row], columns= ['Team ID','GW', 'Points','Total Points','Squad Value', 'Transfers in GW', 'Points on Bench','GW Transfers cost'])
        all_df.append(new_DF)
#         new_DF = new_DF.rename(columns = {"Total Points" : item for item in player_name})
        
        result.append(row)
    
```
