```python
result = []
team_id = data['entry'].tolist()

for team in team_id:
    url_tid = 'https://fantasy.premierleague.com/api/entry/' + str(team) + '/history/'
    p_id = requests.get(url_tid).text
    json_data_p_id = json.loads(p_id)
#     print(json.dumps(json_data_p, indent =2))  #prints JSON for every player
    for x in range(len(json_data_p_id['current'])):
        row = [team, 
               json_data_p_id['current'][x]['event'], 
               json_data_p_id['current'][x]['points'],
               json_data_p_id['current'][x]['total_points'],
               json_data_p_id['current'][x]['value'], 
               json_data_p_id['current'][x]['event_transfers'],
               json_data_p_id['current'][x]['points_on_bench']]
        result.append(row)
```

```python
result
```

```python
player_df = pd.DataFrame(result, columns= ['Team ID','GW', 'Points','Total Points','Sqaud Value', 'Transfers in GW', 'Points on Bench'])
```
